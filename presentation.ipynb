## Read config file + train network

### Dataset 1 (Training > 500) with config 1 (hidden >= 2):

```python
config = utils.read_config_file("config/config_1.txt")

network = Network.Network(config)
network.init_weights()

# If we want to visualize the forward
#network.forward(network.X_train[0], network.y_train[0],True)

network.train(epochs=50)

network.visualize(True,factor=10)
print("Test score:",network.test_score())
```
